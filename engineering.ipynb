```python
import pandas as pd
```

```python
cleanData = pd.read_csv('./data/clean_data.csv')
```

```python
print(cleanData.describe())
```

```python
engData = cleanData.copy()
```

```python
engData['age'].min()
engData['age'].max()
```

```python
# print(engData['Online_Or_Offline'] == 'Online')
print(engData['category'].unique())
```

```python
# One-Hot Encoding
engData['online'] = engData['online_or_offline'] == 'online'
engData['in_store'] = engData['online_or_offline'] == 'in-store'
engData['region_north'] = engData['store_region'] == 'north'
engData['region_south'] = engData['store_region'] == 'south'
engData['region_east'] = engData['store_region'] == 'east'
engData['region_west'] = engData['store_region'] == 'west'
engData['payment_digital'] = engData['payment_method'] == 'digital wallet'
engData['payment_cash'] = engData['payment_method'] == 'cash'
engData['payment_card'] = engData['payment_method'] == 'credit card'
engData['gender_male'] = engData['gender'] == 'male'
engData['gender_female'] = engData['gender'] == 'female'
engData['gender_nb'] = engData['gender'] == 'non-binary'
engData['category_furniture'] = engData['category'] == 'furniture'
engData['category_beauty'] = engData['category'] == 'beauty'
engData['category_clothing'] = engData['category'] == 'clothing'
engData['category_electronics'] = engData['category'] == 'electronics'
engData['category_grocery'] = engData['category'] == 'grocery'
```

```python
bins = [0,25,35,45,55,65,100]
labels = ['18-24','25-34','35-44','45-54','55-64','65+']
engData['age_groups'] = pd.cut(engData['age'], bins=bins, labels=labels)
```

```python
engData['year'] = pd.to_datetime(engData['date']).dt.year
engData['month'] = pd.to_datetime(engData['date']).dt.month
engData['day'] = pd.to_datetime(engData['date']).dt.day
```

```python
# Shorten Non-binary to NA
engData.loc[engData['gender'] == 'Non-binary', 'gender'] = 'N/B'
engData.loc[engData['payment_method'] == 'Credit Card', 'payment_method'] = 'CC'
```

```python
print(engData.head().to_markdown())
```

```python
engData.to_csv('data/eng_data.csv', index_label=False)
```
